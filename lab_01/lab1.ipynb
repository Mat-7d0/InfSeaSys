2. Токенизация Mystem vs razdel.tokenize
Токенизируйте текст (не менее 10 предложений, можно взять любую статью Вики) с помощью razdel и с помощью Mystem. Найдите различия в токенизациях. Что, по вашему мнению, работает лучше на этом тексте?

```python
text = """
111.222 22 33 параметры и опции для настройки сборщика мусора, что позволяет разработчикам оптимизировать его под конкретные требования приложения. Для исследовательских задач по оптимизации сборки мусора в «Julia» был разработан интерфейс программирования приложения (англ. Application programming interface, API), который позволил вынести реализацию сборщика мусора в отдельный модуль и дал возможность независимым разработчикам разрабатывать свои варианты сборщика мусора. Проект «MMTk» в рамках которого разрабатываются различные сборщики мусора недавно объявил о поддержке «Julia» и уже существует порт этого языка программирования для поддержки существующей коллекции собрщиков мусора. Для этого была немного переработана среда выполнения и улучшен API. Для внедрения «MMTk» в различные языки необходимо написать связывающий модуль, который включает специфичный для «MMTk» код мутатора и специфичный для языка программирования «MMTk» код, схема представлена на рисунке 1.1.
"""
```

```python
razdel = [token.text for token in tokenize(text)]
```

```python
print(f'{len(razdel)} токенов')
```

```python
razdel
```

```python
m = Mystem()
```

```python
mystem = [token['text'] for token in m.analyze(text)]
```

```python
print(f'{len(mystem)} токенов')
```

```python
mystem
```

```python
i = j = 0
comparing_tokens = []
unnecessary_tokens = 0

while i < len(razdel) and j < len(mystem):
    if mystem[j] in (' ', '\n'):
        j += 1
        unnecessary_tokens += 1
        continue
    
    print(razdel[i], mystem[j])
    comparing_tokens.append(razdel[i] == mystem[j].strip())
    i += 1
    j += 1

print(i, i == len(razdel), j, j == len(mystem))
print(f'Токенизация{" " if all(comparing_tokens) else " не "}совпадает')
print(f'Количество лишних токенов у mystem {unnecessary_tokens}')
```

Вывод: mystem выделяет и служебные элементы предложения. При нахождении знаков препинания, берет их вместе с пробелом. Лучше использовать razdel.

3. Лемматизация Mystem vs Pymorphy
Лемматизируйте текст с помощью mystem и pymorphy. Найдите различия в лемматизации. Что, по вашему мнению, работает лучше на этом тексте?


```python
m = Mystem()
mystem_lemmas = m.lemmatize(text)
```

```python
morph = MorphAnalyzer()
```

```python
morph_lemmas = []
for mystem_token in mystem:
    if mystem_token in (' ', '\n'):
        continue
    morph_lemmas.append(morph.parse(mystem_token))
```

```python
mystem_lemmas
```

```python
morph_lemmas
```

```python
i = j = k = 0
comparing_lemmas = []
unnecessary_tokens = 0

while i < len(morph_lemmas) or j < len(mystem_lemmas) or k < len(mystem):
    if mystem[k] in (' ', '\n'):
        k += 1
        j += 1
        unnecessary_tokens += 1
        continue
    
    morph_lemma = morph_lemmas[i][0].normal_form
    mystem_lemma = mystem_lemmas[j]
    comparing = morph_lemma.strip() == mystem_lemma.strip()
    if not comparing:
        token = mystem[k]
        print(f'Токен: {token}   Mystem: {mystem_lemma}   PyMorph: {morph_lemma}, ')
        
    comparing_lemmas.append(comparing)
    i += 1
    j += 1
    k += 1

print(i, i == len(razdel), j, j == len(mystem_lemmas), k, k == len(mystem))
print(f'Лемматизация{" " if all(comparing_lemmas) else " не "}совпадает')
print(f'Количество лишних токенов у mystem {unnecessary_tokens}')
```

Вывод: лемматризация работает по разному, pymorph предлагает несколько подходящих значений (сравнивалось наиболее близкое по его мнению). PyMorph берет точную начальную форму.
