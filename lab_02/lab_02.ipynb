```python
from yargy import *
from yargy.predicates import *
from yargy.relations import *
from yargy.interpretation import *
from yargy.tagger import *
from yargy.tokenizer import *
from yargy.pipelines import *

from natasha.grammars.name import NAME
```

Правило: Разбор цитирования

```python
END = rule(type('EOL')).repeatable()
WORD = rule(
    not_(type('EOL'))
)

CITECONTENT = rule(
    type('INT'),
    '.',
    WORD.repeatable().interpretation(
        CiteContent.content
    ),
    END
).interpretation(
    CiteContent
)

cite_parser = Parser(CITECONTENT)

for cite_match in cite_parser.findall(text):
    print(cite_match.fact)

# Разбивка на Ахивировано и т.п.
```

Правило: Каст

```python
Clarification = fact(
    'Clarification',
    ['clarification']
)

ENDOFLINE = rule(type('EOL')).repeatable()
WORD = rule(
    not_(type('EOL'))
)

CLARIFICATION = rule(
    '—',
    WORD.repeatable().interpretation(
        Clarification.clarification
    ),
    ENDOFLINE
).interpretation(
    Clarification
)

Cast = fact(
    'Cast',
    ['name', 'role']
)

CAST = rule(
    NAME.interpretation(
        Cast.name
    ),
    CLARIFICATION.interpretation(
        Cast.role
    )
).interpretation(
    Cast
)

cast_parser = Parser(CAST)

for cast_match in cast_parser.findall(text):
    print(cast_match.fact)

# Странная обработка имен, Америка - имя, но оно игнорируется, а из и Википедия- нормально? ЧЗФ???
```

Правило: Найти правую часть определений

```python
Clarification = fact(
    'Clarification',
    ['clarification']
)

CLAREND = or_(
    rule(eq('.')).optional(),
    rule(type('EOL')).repeatable()
)
CLARWORD = rule(
    and_(
        not_(type('EOL')),
    )
)

CLARIFICATION = rule(
    '—',
    CLARWORD.repeatable().interpretation(
        Clarification.clarification
    ),
    CLAREND
).interpretation(
    Clarification
)

cast_parser = Parser(CLARIFICATION)

for cast_match in cast_parser.findall(text):
    print(cast_match.fact)

# Иногда встречается какой-то управляющий(?) символ,
# например «Ба́рби» (англ. Barbie) — американская, который не обрабатывается,
# не знаю как пофиксить через правила
```

Правило: Продюсеры

```python
COMPANY_NAME = morph_pipeline([
    'Heyday Films',
    'LuckyChap Entertainment',
    'Mattel Films'
])

COPANY_ACTION = fact('Action', ['company', 'action'])

PROF = rule(
    COMPANY_NAME.interpretation(COPANY_ACTION.company),
    gram('VERB').interpretation(COPANY_ACTION.action),
).interpretation(Prof)

cast_parser = Parser(PROF)

for cast_match in cast_parser.findall(text):
    print(cast_match.fact)
```
